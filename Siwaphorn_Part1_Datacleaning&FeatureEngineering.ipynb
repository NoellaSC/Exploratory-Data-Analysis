# Data cleaning

```python
#Drop unwanted data columns
df = df.drop(['Category Id','Customer Email','Customer Password','Customer Zipcode','Product Status','Product Image','Latitude','Longitude','Customer City','Customer State','Customer Street','Order State','Order City','Department Id','Order Item Cardprod Id','Product Card Id','Order Id','Order Item Id','Product Category Id'], axis=1)
df.head()
```

```python
# Drop null value columns
df = df.drop(['Order Zipcode','Product Description'], axis=1)
df.head()
```

```python
# check the results
df.isnull().sum().sum()

# these 8 remaining value are from the Lname as some customers did put their last name
```

```python
#Drop duplicate columns 
df = df.drop(['Order Item Product Price','Sales','Order Profit Per Order','Order Item Total','Order Customer Id'], axis=1)
df.head()
```

# Feature Engineering

```python
#Modify columns
df['Order date'] = pd.to_datetime(df['order date (DateOrders)'])
df['Shipping date']=pd.to_datetime(df['shipping date (DateOrders)'])

df['Order_year'] = pd.DatetimeIndex(df['Order date']).year
df['Shipping_year'] = pd.DatetimeIndex(df['Shipping date']).year
```

```python
# Arrange columns for new data set
new_df = ['Type', 'Days for shipping (real)', 'Days for shipment (scheduled)',
       'Benefit per order', 'Sales per customer', 'Delivery Status',
       'Late_delivery_risk', 'Category Name', 'Customer Country','Customer Id',
       'Customer Fname', 'Customer Lname', 'Customer Segment',
       'Department Name', 'Market', 'Order Country',
       'Order Item Discount', 'Order Item Discount Rate',
       'Order Item Profit Ratio', 'Order Item Quantity', 'Order Region',
       'Order Status', 'Product Name', 'Product Price', 'Shipping Mode',
       'Order_year','Shipping_year']
df=df[new_df]
df
```
